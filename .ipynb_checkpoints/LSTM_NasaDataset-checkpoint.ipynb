```python
def add_remaining_useful_life(df):
    grouped_by_unit = df.groupby(by="id")
    max_cycle = grouped_by_unit["time_cycles"].max()
    
    result_frame = df.merge(max_cycle.to_frame(name='max_cycle'), left_on='id', right_index=True)

    remaining_useful_life = result_frame["max_cycle"] - result_frame["time_cycles"]
    result_frame["RUL"] = remaining_useful_life

    result_frame = result_frame.drop("max_cycle", axis=1)
    return result_frame
  
train = add_remaining_useful_life(train)
train[index_names+['RUL']].head()
```

```python
def remove_low_var_features(df, features, threshold):
    low_var_features = []
    low_var_values = []
    non_low_var=[]

    for f in features:
        
        std=train[f].describe().loc['std']
        
        var = std*std

        if var <= threshold:
            low_var_features.append(f)
            low_var_values.append(var)
            
            print(f + ': var: ' + str(var))
            df=df.drop(f,axis=1)
        else:
            non_low_var.append(f)
    
    return df, low_var_features, low_var_values,non_low_var
```

```python
from sklearn.metrics import mean_squared_error, r2_score
def evaluate(y_true, y_hat, label='test'):
    mse = mean_squared_error(y_true, y_hat)
    rmse = np.sqrt(mse)
    variance = r2_score(y_true, y_hat)
    print('{} set RMSE:{}, R2:{}'.format(label, rmse, variance))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
features_set, labels = np.array(train[sensor_names]), np.array(train['RUL'])
features_set.shape
```

```python
import tensorflow as tf
features_set = np.reshape(features_set, (features_set.shape[0],9))
features_set=tf.keras.layers.RepeatVector(4)(features_set)
features_set.shape
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(4,9)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dense(units = 1))
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
engine_list=[76,81]

for i in engine_list:
    a=str('id == '+str(i))
    temp_df=test.query(a).copy()
    test_features_set, labels = np.array(temp_df[sensor_names]), np.array(test['RUL'])
    test_features_set = np.reshape(test_features_set, (test_features_set.shape[0],9))
    test_features_set=tf.keras.layers.RepeatVector(4)(test_features_set)
    predictions = model.predict(test_features_set)
    predictions_df=pd.DataFrame(predictions, columns = ['RUL'])
    
    print("Engine #%d" % i)
    temp_df=temp_df.reset_index(drop=True)
    plt.figure(figsize=(10,6))
    plt.plot(temp_df['RUL'], color='blue', label='Test RUL')
    plt.plot(predictions_df, color='red', label='Predicted RUL')
    plt.title('RUL Prediction')
    plt.ylabel('RUL')
    plt.legend()
    plt.show()
    
    evaluate(temp_df['RUL'],predictions_df,'RUL')
    
    table=pd.DataFrame()
    table['Actual']=temp_df['RUL']
    table['Predicted']=predictions_df['RUL']
    #pd.set_option('display.max_rows',table.shape[0]+1)
    display(table)
```

```python
train['nowarning']=np.where(train['label1']==0,1,0)
train['warning']=np.where(train['label1']==1,1,0) 
warninglabels=['nowarning','warning']
train
```

```python
features_set2, labels2 = np.array(train[sensor_names]), np.array(train[warninglabels])
features_set.shape

features_set2 = np.reshape(features_set2, (features_set2.shape[0],9))
features_set2=tf.keras.layers.RepeatVector(4)(features_set2)


model = Sequential()

model.add(LSTM(units=50, return_sequences=True, recurrent_activation='hard_sigmoid' ,input_shape=(4,9)))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=True,recurrent_activation='hard_sigmoid'))
model.add(Dropout(0.2))

model.add(LSTM(units=50, recurrent_activation='hard_sigmoid'))

model.add(Dense(units = 2,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(features_set2, labels2, epochs = 50, batch_size = 16)
```

```python
import seaborn as sns
from sklearn import metrics
test['nowarning']=np.where(test['label1']==0,1,0)
test['warning']=np.where(test['label1']==1,1,0) 

engine_list=[81,91]
for i in engine_list:
    print("\nEngine #%d" %i)
    a=str('id == '+str(i))
    temp_df=test.query(a).copy()
    test_features_set, labels = np.array(temp_df[sensor_names]), np.array(temp_df[warninglabels])
    test_features_set = np.reshape(test_features_set, (test_features_set.shape[0],9))
    test_features_set=tf.keras.layers.RepeatVector(4)(test_features_set)
    y_pred=model.predict(test_features_set)
    df = pd.DataFrame(y_pred, columns = ['nowarning','warning'])
    df['label']=np.where(df['nowarning']>df['warning'],0,1)
    pd.set_option('display.max_rows',df.shape[0]+1)
    con_mat = tf.math.confusion_matrix(labels=temp_df['label1'], predictions=df['label']).numpy()
    print("Confusion Matrix:")
    print(con_mat)
    print("Accuracy:",metrics.accuracy_score(temp_df['label1'], df['label']))
    print("Precision:",metrics.precision_score(temp_df['label1'], df['label'],average='macro'))
    print("Recall: ",metrics.recall_score(temp_df['label1'], df['label'],average='macro'))
```
